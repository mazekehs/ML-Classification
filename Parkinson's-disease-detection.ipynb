# Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score

import warnings
warnings.filterwarnings("ignore")
```

# Data Collection and Analysis

```python
#Loading the data from csv file to pandas dataframe

parkinsons_data=pd.read_csv('/kaggle/input/parkinsons-data-set/parkinsons.data')
```

```python
parkinsons_data.head()
```

```python
#number of rows and columns

parkinsons_data.shape
```

```python
#Data Information

parkinsons_data.info()
```

```python
#checking for missing values in each column

parkinsons_data.isnull().sum()
```

```python
#getting the statistical measures from the data

parkinsons_data.describe()
```

```python
#distribution of target value--here 'status'

parkinsons_data['status'].value_counts()
```

1 -> Parkinson's Positive

0 ->Healthy

```python
#grouping the data based on the target value

parkinsons_data.groupby('status').mean(numeric_only=True)
```

# Data preprocessing 

```python
#Separating features and target

X=parkinsons_data.drop(columns=['name','status'],axis=1)
Y=parkinsons_data['status']
```

```python
print(X)
```

```python
print(Y)
```

# Splitting the data to training data and Test data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(Y.shape,Y_train.shape,Y_test.shape)
```

# Data Standardization

```python
scaler=StandardScaler()
```

```python
scaler.fit(X_train) #should never fit X_test
```

```python
X_train=scaler.transform(X_train)

X_test=scaler.transform(X_test)
```

```python
print(X_train)
```

# Model training

**Support Vector Machine**

```python
model=svm.SVC(kernel='linear')
```

```python
#training the SVM model with training data
model.fit(X_train,Y_train)
```

# Model evaluation

**Accuracy Score**

```python
#accuracy score on training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(Y_train,X_train_prediction)
```

```python
print("Accuracy of training data :",training_data_accuracy)
```

```python
#accuracy score on testing data
X_test_prediction=model.predict(X_test)
testing_data_accuracy=accuracy_score(Y_test,X_test_prediction)
```

```python
print("Accuracy of testing data :",testing_data_accuracy)
```

# Building a predictive system

```python
input_data=(119.99200,157.30200,74.99700,0.00784,0.00007,0.00370,0.00554,0.01109,0.04374,0.42600,0.02182,0.03130,0.02971,0.06545,0.02211,21.03300,0.414783,0.815285,-4.813031,0.266482,2.301442,0.284654)

#changing input data to numpy array
input_data_array=np.asarray(input_data)

#reshape the numpy array
input_data_reshaped=input_data_array.reshape(1,-1)

#standardize the data

std_data=scaler.transform(input_data_reshaped)

prediction=model.predict(std_data)
print(prediction)

if(prediction[0]==0):
    print("The Person does not have Parkinsons Disease")
else:
    print("The Person has Parkinsons Disease")
    
```
